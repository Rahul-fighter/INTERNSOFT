```
#Classes
#How not to create a class
class Car:
    pass
psych_car = Car()
psych_car.name = 'Maruthi'
psych_car.color = 'Red' 
psych_car.mfg_date = 2000
print(psych_car)
print("My %s is %s in color, it's a %d model." %(psych_car.name,psych_car.color,psych_car.mfg_date))
```

```
#How to actually create a class
class Car:
    def __init__(self,name,color,mfg_date):
        self.name = name
        self.color = color
        self.mfg_date = mfg_date

psych_car = Car('Maruthi','Red',2000)
print(psych_car)
print("My %s is %s in color, it's a %d model." %(psych_car.name,psych_car.color,psych_car.mfg_date))
```

```
#Methods in a class
class Car:
    def __init__(self,name,color,mfg_date):
        self.name = name
        self.color = color
        self.mfg_date = mfg_date

    def age_of_car(self,curr_date):
        age = curr_date - self.mfg_date
        return "It's %d years old now." %age

    def __str__(self):
        return "My %s is %s in color, it's a %d model." %(psych_car.name,psych_car.color,psych_car.mfg_date)

psych_car = Car('Maruthi','Red',2000)
print(psych_car)
psych_car.age_of_car(2020)
```

```
#Inheritence
class Engine(Car):
    def __init__(self,engine_type,*args,**kwargs):
        super(Engine,self).__init__(*args,**kwargs)
        self.engine_type = engine_type

    def __str__(self):
        return "My %s is %s in color, it's a %d model and it has a %s engine" %(psych_car.name,psych_car.color,psych_car.mfg_date,self.engine_type)

psych_engine = Engine('IC Engine','Maruthi','Red',2000)
print(psych_engine)
psych_engine.age_of_car(2020)
print(isinstance(psych_engine,Car))
print(isinstance(psych_engine,object))
```

```
#Polymorphism property
class Mul():
    def __init__(self,a,b):
        self.a = a
        self.b = b
    
    def disp(self):
        return self.a*self.b

psych_nos = Mul(4,5)
psych_nos1 = Mul('a',5)
psych_nos2 = Mul([1,2,3],4)
print(psych_nos.disp())
print(psych_nos1.disp())
print(psych_nos2.disp())
```
