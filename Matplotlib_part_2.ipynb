```
#Line plots(colors and styles)
import numpy as np
import pandas as pd
import matplotlib as mlp
import matplotlib.pyplot as plt
n = np.linspace(0,20,200)
plt.plot(n,np.cos(n-1),color = 'blue')
plt.plot(n,np.cos(n-3),color = 'g')
plt.plot(n,np.cos(n-4),color = '0.75')
```

```
#Using different linestyle and adjusting plot limit
plt.plot(n,np.cos(n-1),linestyle = ':')
plt.plot(n,np.cos(n-3),linestyle = '-.')
plt.plot(n,np.cos(n-4),linestyle = '--')
```

```
# Adjusting plot limit
plt.plot(n,np.sin(n),color = '0.2255',linestyle = '--')
plt.xlim(1,12)
plt.ylim(-2,2)
```

```
#Labeling plots
plt.plot(n,np.sin(n),linestyle ="-.",color='y')
plt.title("Sinusiodal wave")
plt.xlabel('x axis')
plt.ylabel('y axis')
```

```
#Scatter plots
plt.style.use('seaborn-whitegrid')
x = np.linspace(0,20,30)
y = np.log(x) 
z = np.pi*x
plt.plot(x,y,'o-',color='red')
plt.plot(x,z,'ob')

```

```
#Using Scatter function
plt.scatter(x,y,marker = '+')
```

```
#Use of scatter fumc
state = np.random.RandomState(5)
a = state.randn(50)
b = state.randn(50)
colors = state.randn(50)
size = 1000*state.randn(50)
plt.scatter(a,b,c = colors,s= size,alpha = 0.25, cmap = 'winter')
plt.colorbar()
```

```
#Errorbars
s = np.linspace(0,20,75)
er = 0.275
m = np.cos(s) + er*np.random.randn(75)
plt.errorbar(s,m,xerr=er,fmt = 'none',color = 'y')
```

```
#Histograms
info = np.random.randn(45000)
plt.hist(info)
```

```
#More controlled display
plt.hist(info, bins = 7, alpha = .75, histtype = 'stepfilled')
```

```
#Two dimensional histogram
plt.hist2d(x,y,bins = 50, cmap = 'summer')
plt.colorbar()
```

```
#Hexagonal bins
plt.hexbin(x,y,gridsize = 20,cmap = 'BuPu')
cb = plt.colorbar()
```
