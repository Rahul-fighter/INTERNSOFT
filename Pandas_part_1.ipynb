```
#Panda version
import pandas
print(pandas.__version__)
import numpy as np 
import pandas as pd
```

```
#Series
hellos = pd.Series([220, 625, 420, 10])
print(hellos)
print(hellos.values)
hellos.index
```

```
#Examples
fruits = pd.Series([10, 20, 30, 40], 
    index=['Banana', 'Apple', 'Orange', 'Grapes'])
print(fruits)
print(fruits[2])
print(fruits['Grapes'])
```

```
#Naming the values and indexes
fruits.name = "Type"
fruits.index.name = "Quantity"
fruits
```

```
#Numpy operations on series
print(np.sin(fruits))
print("")
print(np.pi*fruits)
print("")
print(np.cos(fruits))
print("")
fruits[fruits>30]
```

```
#Creating series from a dictionary
ICC_ranking = {1:"India",2:"Australia",3:"South Africa",4:"England",5:"New Zealand",6:"West Indies"}
print(ICC_ranking)
print("")
ICC_ranking_1  = pd.Series(ICC_ranking)
print(ICC_ranking_1)
print("")
ICC_ranking_1.index.name = "Position"
ICC_ranking_1.name = "Country"
print(ICC_ranking_1)
```

```
#Dataframes
class_room = {"Names":["Cheta","Sade","Question_boy","Mutual_dove","Deeps"],
              "Sem1_marks":[85,87,99,92,100],
              "Sem2_marks":[87,85,45,90,100]}
print(class_room)
print("")
cl_room = pd.DataFrame(class_room)
print(cl_room)
print("")
cl_room["Average"] = (cl_room.Sem1_marks + cl_room.Sem2_marks)/2
print(cl_room)
print("")
cl_room["Gender"] = pd.Series(["M","M","M","F","F"])
cl_room
```

```
#Conversions
print(cl_room.to_dict())
print('''
       ''')
pd.DataFrame(cl_room)
cl_room
```

```
#Creating dataframes from a list of dicts
squares_and_cubes = [{"x":i**2,"y":i**3}for i in range(10)]
print(squares_and_cubes)
print('''

      ''')
pd.DataFrame(squares_and_cubes)
```
