```
#Index Alignments
High_score = pd.Series({'Sachin':99,'Ganguly':97,
                    'Sehwag':100,'Yuvraj':96},
                     name='players')
Average = pd.Series( {'Sehwag':56,'Yuvraj':65,
                    'Dhoni':71,'Kholi':72},name = 'average')
print(High_score)
print(Average)
print(High_score+Average)
print(High_score.index | Average.index)
```

```
#Multiplying data of diff series
data_1 = pd.Series(list(range(6)),index = [0,1,2,3,4,5])
data_2 = pd.Series([10,30,20,50,40,60],index = [1,2,3,4,5,6])
print(data_1)
print("")
print(data_2)
print("")
print(data_1*data_2)
print("")
data_1.add(data_2,fill_value=0)
```

```
#Merge operation
#Dataframes
class_room = {"Names":["Cheta","Sade","Question_boy","Mutual_dove","Deeps"],
              "Sem1_marks":[85,87,99,92,100],
              "Sem2_marks":[87,85,45,90,100]}
cl_room = pd.DataFrame(class_room)
cl_room["Section"] = pd.Series(["A","B",'B','C','A'])
cl_room["Average"] = (cl_room.Sem1_marks + cl_room.Sem2_marks)/2
cl_room["Gender"] = pd.Series(["M","M","M","F","F"])
cl_room
```

```
#Continuation
cl_room2 = pd.DataFrame({"Section":["A","B",'C'],'Sem3_marks' : [90,91,93]})
print(cl_room2)
print('''
      ''')
cl_room.merge(cl_room2)
```
