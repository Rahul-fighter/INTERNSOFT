```
#Importong matplotlib
import numpy as np
import pandas as pd
import matplotlib as mlp
import matplotlib.pyplot as plt
mlp.__version__
```

```
#Alpha plot
plt.plot(np.random.normal(0,10,50),np.random.normal(0,20,50),'c')
```

```
#Using different shape
%matplotlib inline
plt.plot(np.random.normal(0,10,50),np.random.normal(0,20,50),'s')
```

```
#Plotting in pandas
aye = pd.Series(np.random.normal(size = 22))
print(aye)
print(aye.cumsum(0).plot(grid =  True))
plt.plot(aye,'o')
```

```
#Plotting a dataframe
distributions = pd.DataFrame({'normal':np.random.normal(size = 20),
                 'gamma':np.random.gamma(1, size = 20),
                 'poisson':np.random.poisson(size = 20)})
print(distributions)
distributions.cumsum(0).plot()
distributions.cumsum(0).plot(subplots = True)
distributions.cumsum(0).plot(secondary_y = 'gamma')
```

```
#More controlled display
f,a = plt.subplots(ncols= 3, nrows= 1, figsize=(20,3))
for i,var in enumerate(['normal','gamma','poisson']):
    distributions[var].cumsum(0).plot(ax = a[i], title = var)
a[0].set_ylabel('values')
```

```
#Saving figures
f.savefig('my_fig.jpg')
from IPython.display import Image
print(f.canvas.get_supported_filetypes())
Image('my_fig.jpg')
```
