```
# Different ways of declaring strings
a = 'people,'
b = 'Hi there'
c = "Let's begin"
print(b,a,c)
```

```
#Printing string
print('Hola amigo')
'Como estas'
```

```
#Finding length of string
x="Good day"
len(x)
```

```
#Accessing strings using index
t = "Stealth"
print(t[0],t[1],t[2],t[3],t[4],t[5],t[6])
t[3]
```

```
#Slicing strings
s = 'Bohemian Raphsody'
print(s[:])
print(s[1:4])
print(s[:3])
s[::2]
```

```
#To access a string using negative index
print(s[-1])
s[:-1]
```

```
#To print a string backwords
s[::-1]
```

```
#Concatenation and reassigning strings
u = 'Kem chho? '
v = 'Maja ma? '
print(u+v)
u = u+v
u
```

```
#Repition of strings
u*5
```

```
#Built in string methods
m = 'Heavens fall'
print(m.lower())
print(m.upper())
print(m.split())
m.split('s')
```

```
#Print formatting
"The  two top musicians in the world are {} and {}".format('Eminem','Freddie Mercury') 
```

```
#Location and Counting
print(m)
print(m.count('e'))
m.find('l')
```

```
#Formatting strings
print(m.center(40,'f'))
"Good\tmorning".expandtabs()
```

```
#Using 'is' checks
n = "naman"
print(n.isalnum())
print(n.isalpha())
print(n.islower())
print(n.isupper())
print(n.isspace())
print(n.istitle())
n.endswith(n)

```

```
#Built in reg expression
print(n.split('a'))
print(n.partition('a'))
```

```
# List functions,indexing and slicing
vowels = ['a','e','i','o','u']
print(vowels)
print(len(vowels))
print(vowels[0])
print(vowels[1:3])
```

```
# Concatenating and reassigning list
room = ['bed','chair','light']
room + ['screen']
print(room)
room = room + ['screen']
room
```

```
# Append, Pop 
alp = ['a','b','d','c']
alp.append('e')
print(alp.pop(0))
alp
```

```
#Reverse and sort
num = [4,5,9,8,7]
num.reverse()
print(num)
num.sort()
print(num)
```

```
#Nesting lists
R1 = [4,5,6]
R2 = [3,8,1]
R3 = [2,7,9]
Mat = [R1,R2,R3]
Mat
```

```
#accessing elements of mat
print(Mat[2])
Mat[2][1]
```

```
#Building a list comprehension
col2 = [row[2] for row in Mat]
col2
```

```
#advanced lists
list_it = [9,8,4,5,4]
list_it.append(2)
print(list_it)
print(list_it.count(4))
list_it.extend([0,4])
print(list_it)
print(list_it.index(4))
list_it.insert(4,'a')
print(list_it)
ab = list_it.pop()
print(ab)
list_it.remove('a')
print(list_it)
list_it.reverse()
print(list_it)
list_it.sort()
print(list_it)
```
