```
#Program for prime number
import math
def prime_no(no):
    if (no%2==0 and no > 2):
        return False
    for i in range(3,int(math.sqrt(no))+1,2):
        if no % i == 0:
            return False
    return True
print(prime_no(2))
prime_no(12)
```

```
#Iterators and Generators
#Function for square root
def square_rt(num):
    for num in range(6):
       yield num**2
print(list(square_rt(7)))
for x in square_rt(8):
   print(x)
```

```
#Function for fibonnaci series
def fib_dude(n):
     a = 1
     b = 1
     for i in range(n):
        yield a
        a,b = b,a+b

for i in fib_dude(12):
    print(i)
```

```
#Next fuction
def next_gen(n):
    for i in range(n):
        yield i
x = next_gen(5)
print(next(x))
print(next(x))
print(next(x))
print(next(x))

```

```
#Iter function
f = "wassup"
i_f = iter(f)
print(next(i_f))
print(next(i_f))
print(next(i_f))
print(next(i_f))
```

```
#Map function
import  math
def power_p(x):
  return x**2
def root_r(x):
  return int(math.sqrt(x))

tic = [2,3,4,5]

f_power = list(map(power_p,tic))
print(f_power)
f_root = list(map(root_r,f_power))
f_root
```

```
#Using lambda
print(list(map(lambda x:x**2 ,tic)))
list(map(lambda x,y:x+y ,tic,f_power))
```

```
#Using reduce function
from functools import reduce
red_test = [1,2,3,4,5,6]
reduce(lambda x,y:x*y , red_test)
```
