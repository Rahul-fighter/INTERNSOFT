```
#Using finally block
def is_float():
    try:
        val = float(input("Please enter the input : "))
    except:
        print("Error:The input is not float")
    finally:
        print("Done deal")
    print(val)
is_float()
```

```
#To disp the false datatype
def is_float():
    try:
        val = float(input("Please enter the input : "))
    except:
        print("Error:The input is not float")
        val = float(input("Try again : "))
    finally:
        print("Done deal")
    print(val)
is_float()
```

```
#To continously take values
def is_float():
  while True:
    try:
        val = float(input("Please enter the input : "))
    except:
        print("Error:The input is not float")
        continue
    else:
        print("You've got it")
        break
    finally:
        print("Done deal")
    print(val)
is_float()
```
