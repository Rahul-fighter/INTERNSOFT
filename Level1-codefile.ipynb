```
#For loop
#program to find sum of all even numbers
group = [1,2,3,4,5,6,7,8]
sum=0;
for i in group:
    if i%2==0:
        sum+=i
print(sum)
```

```
#While loop
#Program to print sum of all odd no
#In given range
x = int(input("Enter the range "))
sum=0
i=0
while i<x:
    if i%2 != 0:
        sum+=i;
    i=i+1
print(sum)
print("end")
```
